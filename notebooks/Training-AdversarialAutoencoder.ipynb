# Imports and Functions

```python
from sunlab.environment.base.cuda import *
from sunlab.environment.base.extras import *
from sunlab.globals import FILES
from sunlab.sunflow import *
```

# Runtime Parameters

```python
SAVE_EXTRAS = False
```

```python
g_Base = lambda *args: DIR_ROOT + "Figures/AdversarialAutoencoder/"
```

# Setup

## Model

```python
MODEL_FOLDER = DIR_ROOT + "models/adversarial_autoencoder/"
DISTRIBUTION = GaussianDistribution
```

```python
model, dataset = create_aae_and_dataset(FILES['TRAINING_DATASET'], MODEL_FOLDER, normalization_scaler=MaxAbsScaler, distribution=DISTRIBUTION, shuffle=True, val_split=0.1)
```

## Epochs

```python
_epochs = 1600
```

## Training

```python
_ = model.train(dataset, _epochs, output=True, output_freq=100)
```

# Plots

```python
plt.scatter2d(model.encoder(dataset.dataset), cmap=Pmap, c=dataset.labels, s=1)
if SAVE_EXTRAS:
     plt.save(g_Base() + "LatentSpace_" + model.distribution.get_name())
```
