## Cumulative Morphological Latent Spaces

```python
sns.jointplot(training_data.dataframe, x='Latent-0', y='Latent-1', hue='Class', s=2, palette=Pmap)
plt.gca().get_legend().remove()
plt.apply_boundary()
if SAVE_EXTRAS:
    plt.save(g_Base() + "Cumulative_24hr")
```
